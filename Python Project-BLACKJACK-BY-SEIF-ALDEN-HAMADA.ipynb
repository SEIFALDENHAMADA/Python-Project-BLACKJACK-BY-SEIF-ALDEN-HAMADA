```python
# let's have fun------------------

import random as ra
import time as t
class Base:
    dealerNamelist = ["AHMED", "SEIF", "ABDO", "OMAR", "YOUSIF"]
    Dealer_name = ra.choice(dealerNamelist)
    cards = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'K', 'Q', 'J', 'A']

    def __init__(self, Player_name):
        self.Player_name = Player_name

    def point_calculation(self, cards):
        points = 0
        aces = 0
        for card in cards:
            if card in "KQJ":
                points += 10
            elif card == "A":
                aces += 1
            else:
                points += int(card)  
        for i in range(aces):
            if (points + 11) > 21:
                points += aces
            else:
                points += 11    

        return points
    


class Game(Base):
    def __init__(self, Player_name):
        super().__init__(Player_name)

    def Playing(self):
        cash = 300
        round = 1
        while True:
            print("Welcome To Our Super BlackJack Game..")
            t.sleep(0.5)
            print("Choose The Difficulty By Number")
            t.sleep(0.5)
            print("1/ Choose The HARD ")
            print("2/ Choose The MED ")
            print("3/ Choose The EASY ")
            diff = int(input("CHOOSE YOUR HARDNESS: "))
            if diff == 1:
                t.sleep(0.5)
                print("Difficulty = HARD ")
                while cash != 0:
                    Dealer_spend = ra.choice([50, 30, 20, 35, 40 , 45, 35, 55,60,70,65,75])
                    player_cards = [ra.choice(self.cards) for i in range(2)]
                    Dealer_Cards = [ra.choice(self.cards) for i in range(2)]
                    t.sleep(0.5)
                    print("---------------------------------------------------------------------")
                    t.sleep(0.5)
                    print("Hello, let's start playing... GAME DEVELOPED BY SEIF ALDEN HAMADA")
                    t.sleep(0.5)
                    print(f"Hello M.R: {self.Player_name}, This is M.R {self.Dealer_name} I am The Dealer and I am With you")
                    t.sleep(0.5)
                    print(f"ROUND: {round}, CASH: {cash}$")
                    t.sleep(0.5)
                    spend_cash = int(input(f"Pick Your Betting Amount, Minimum Amount Is 50$, Your Cash Balance is '{cash}$' Take Care: "))
                    if spend_cash < 50 and cash >= 50:
                        spend_cash = 50
                        t.sleep(0.5)
                        print(f" Sorry This Amount Below The Minimum Amount, So You Spent {spend_cash}$")
                        cash -= spend_cash
                        pass
                    elif spend_cash > cash or spend_cash < 50:
                        t.sleep(0.5)
                        print("Sorry The Minimum Amount Is Higher Than Your Cash :(")
                        break   
                    
                        
                    print(f"You Spent {spend_cash}$")
                    t.sleep(0.5)
                    print(f"Dealer Spent {Dealer_spend}$")
                    t.sleep(0.5)
                    print(f"My cards are (Dealer cards): ( {Dealer_Cards[0]} )( * )")
                    t.sleep(0.5)
                    print(f"Your cards are (Player cards): ( {player_cards[0]} ) ( {player_cards[1]} )")

                    player_points = self.point_calculation(player_cards)
                    dealer_cards_points = self.point_calculation(Dealer_Cards)
                    t.sleep(0.5)
                    print(f"Total = {player_points}")

                    while True:
                        print("_____HIT_______STAND________HIT______STAND_______HIT________STAND________HIT")
                        hitOrno = input("Do You Want To Hit y/n? ").lower()
                        if hitOrno == "y":
                            new_card = ra.choice(self.cards)
                            player_cards.append(new_card)
                            player_points = self.point_calculation(player_cards)
                            print(f"You drew: {new_card}")
                            t.sleep(0.5)
                            print(f"Your cards are: {player_cards} (Total: {player_points})")
                            if player_points > 21:
                                print("Buuuuuuust! You exceeded 21. Dealer wins.")
                                cash -= spend_cash
                                print(f"{cash + spend_cash }-{spend_cash}$ = You Current Balance is {cash}$")
                                round += 1

                                return
                        else:
                            break

                    t.sleep(0.5)    
                    print(f"Dealer's cards are: {Dealer_Cards}")
                    while dealer_cards_points < 17:
                        new_card = ra.choice(self.cards)
                        Dealer_Cards.append(new_card)
                        dealer_cards_points = self.point_calculation(Dealer_Cards)
                        t.sleep(0.5)
                        print(f"Dealer drews: {new_card}")

                    t.sleep(0.5)
                    print(f"Dealer's final cards: {Dealer_Cards} (Total: {dealer_cards_points})")

                    if dealer_cards_points > 21 or player_points > dealer_cards_points:
                        t.sleep(0.5)
                        print("Congratulations! You win!")
                        cash += Dealer_spend
                        t.sleep(0.5)
                        print(f"{cash - Dealer_spend} + {Dealer_spend}$ = You Current Balance is {cash}$")
                        round += 1

                    elif dealer_cards_points > player_points:
                        t.sleep(0.5)
                        print("Dealer wins. Better luck next time!")
                        cash -= spend_cash
                        t.sleep(0.5)
                        print(f"{cash + spend_cash }-{spend_cash}$ = You Current Balance is {cash}$")
                        round += 1

                    else:
                        t.sleep(0.5)
                        print("It's a Draaaaaaw")
                        t.sleep(0.5)
                        print(f"You Current Balance is {cash - spend_cash}$ + {spend_cash}$ = TOTAL = {cash}")
                        round += 1
                else:
                    t.sleep(0.5)
                    print("You don't have Enough Money Good BYEEEEE")  
                break           
            else:
                t.sleep(0.5)
                print("Invalid Number")        
            if diff == 2:
                t.sleep(0.5)
                print("Difficulty = MED ")
                while cash != 0:
                    Dealer_spend = ra.choice([50, 30, 20, 35, 40 , 45, 35, 55,60,70,65,75])
                    player_cards = [ra.choice(self.cards) for i in range(2)]
                    Dealer_Cards = [ra.choice(self.cards) for i in range(2)]
                    t.sleep(0.5)
                    print("---------------------------------------------------------------------")
                    t.sleep(0.5)
                    print("Hello, let's start playing... GAME DEVELOPED BY SEIF ALDEN HAMADA")
                    t.sleep(0.5)
                    print(f"Hello M.R: {self.Player_name}, This is M.R {self.Dealer_name} I am The Dealer and I am With you")
                    t.sleep(0.5)
                    print(f"ROUND: {round}, CASH: {cash}$")
                    t.sleep(0.5)
                    spend_cash = int(input(f"Pick Your Betting Amount, Minimum Amount Is 30$, Your Cash Balance is '{cash}$' Take Care: "))
                    if spend_cash < 30 and cash >= 30:
                        spend_cash = 30
                        t.sleep(0.5)
                        print(f" Sorry This Amount Below The Minimum Amount, So You Spent {spend_cash}$")
                        cash -= spend_cash
                        pass
                    elif spend_cash > cash or spend_cash < 30:
                        t.sleep(0.5)
                        print("Sorry The Minimum Amount Is Higher Than Your Cash :(")
                        break   
                    
                        
                    print(f"You Spent {spend_cash}$")
                    t.sleep(0.5)
                    print(f"Dealer Spent {Dealer_spend}$")
                    t.sleep(0.5)
                    print(f"My cards are (Dealer cards): ( {Dealer_Cards[0]} )( * )")
                    t.sleep(0.5)
                    print(f"Your cards are (Player cards): ( {player_cards[0]} ) ( {player_cards[1]} )")

                    player_points = self.point_calculation(player_cards)
                    dealer_cards_points = self.point_calculation(Dealer_Cards)
                    t.sleep(0.5)
                    print(f"Total = {player_points}")

                    while True:
                        print("_____HIT_______STAND________HIT______STAND_______HIT________STAND________HIT")
                        hitOrno = input("Do You Want To Hit y/n? ").lower()
                        if hitOrno == "y":
                            new_card = ra.choice(self.cards)
                            player_cards.append(new_card)
                            player_points = self.point_calculation(player_cards)
                            print(f"You drew: {new_card}")
                            t.sleep(0.5)
                            print(f"Your cards are: {player_cards} (Total: {player_points})")
                            if player_points > 21:
                                print("Buuuuuuust! You exceeded 21. Dealer wins.")
                                cash -= spend_cash
                                print(f"{cash + spend_cash }-{spend_cash} = You Current Balance is {cash}$")
                                round += 1

                                return
                        else:
                            break

                    t.sleep(0.5)    
                    print(f"Dealer's cards are: {Dealer_Cards}")
                    while dealer_cards_points < 17:
                        new_card = ra.choice(self.cards)
                        Dealer_Cards.append(new_card)
                        dealer_cards_points = self.point_calculation(Dealer_Cards)
                        t.sleep(0.5)
                        print(f"Dealer drews: {new_card}")

                    t.sleep(0.5)
                    print(f"Dealer's final cards: {Dealer_Cards} (Total: {dealer_cards_points})")

                    if dealer_cards_points > 21 or player_points > dealer_cards_points:
                        t.sleep(0.5)
                        print("Congratulations! You win!")
                        cash += Dealer_spend
                        t.sleep(0.5)
                        print(f"{cash - Dealer_spend} + {Dealer_spend}$ = You Current Balance is {cash}$")
                        round += 1

                    elif dealer_cards_points > player_points:
                        t.sleep(0.5)
                        print("Dealer wins. Better luck next time!")
                        cash -= spend_cash
                        t.sleep(0.5)
                        print(f"{cash + spend_cash }-{spend_cash} = You Current Balance is {cash}$")
                        round += 1

                    else:
                        t.sleep(0.5)
                        print("It's a Draaaaaaw")
                        t.sleep(0.5)
                        print(f"You Current Balance is {cash - spend_cash}$ + {spend_cash}$ = TOTAL = {cash}")
                        round += 1
                else:
                    t.sleep(0.5)
                    print("You don't have Enough Money Good BYEEEEE")    

                break            
            else:
                t.sleep(0.5)
                print("Invalid Number")
                            
            if diff == 3:
                t.sleep(0.5)
                print("Difficulty = EASY ")
                while cash != 0:
                    Dealer_spend = ra.choice([50, 30, 20, 35, 40 , 45, 35, 55,60,70,65,75])
                    player_cards = [ra.choice(self.cards) for i in range(2)]
                    Dealer_Cards = [ra.choice(self.cards) for i in range(2)]
                    t.sleep(0.5)
                    print("---------------------------------------------------------------------")
                    t.sleep(0.5)
                    print("Hello, let's start playing... GAME DEVELOPED BY SEIF ALDEN HAMADA")
                    t.sleep(0.5)
                    print(f"Hello M.R: {self.Player_name}, This is M.R {self.Dealer_name} I am The Dealer and I am With you")
                    t.sleep(0.5)
                    print(f"ROUND: {round}, CASH: {cash}$")
                    t.sleep(0.5)
                    spend_cash = int(input(f"Pick Your Betting Amount, Minimum Amount Is 15$, Your Cash Balance is '{cash}$' Take Care: "))
                    if spend_cash < 15 and cash >= 15:
                        spend_cash = 15
                        t.sleep(0.5)
                        print(f" Sorry This Amount Below The Minimum Amount, So You Spent {spend_cash}$")
                        cash -= spend_cash
                        pass
                    elif spend_cash > cash or spend_cash < 15:
                        t.sleep(0.5)
                        print("Sorry The Minimum Amount Is Higher Than Your Cash :(")
                        break   
                    
                        
                    print(f"You Spent {spend_cash}$")
                    t.sleep(0.5)
                    print(f"Dealer Spent {Dealer_spend}$")
                    t.sleep(0.5)
                    print(f"My cards are (Dealer cards): ( {Dealer_Cards[0]} )( * )")
                    t.sleep(0.5)
                    print(f"Your cards are (Player cards): ( {player_cards[0]} ) ( {player_cards[1]} )")

                    player_points = self.point_calculation(player_cards)
                    dealer_cards_points = self.point_calculation(Dealer_Cards)
                    t.sleep(0.5)
                    print(f"Total = {player_points}")

                    while True:
                        print("_____HIT_______STAND________HIT______STAND_______HIT________STAND________HIT")
                        hitOrno = input("Do You Want To Hit y/n? ").lower()
                        if hitOrno == "y":
                            new_card = ra.choice(self.cards)
                            player_cards.append(new_card)
                            player_points = self.point_calculation(player_cards)
                            print(f"You drew: {new_card}")
                            t.sleep(0.5)
                            print(f"Your cards are: {player_cards} (Total: {player_points})")
                            if player_points > 21:
                                print("Buuuuuuust! You exceeded 21. Dealer wins.")
                                cash -= spend_cash
                                print(f"{cash + spend_cash }-{spend_cash} = You Current Balance is {cash}$")
                                round += 1

                                return
                        else:
                            break

                    t.sleep(0.5)    
                    print(f"Dealer's cards are: {Dealer_Cards}")
                    while dealer_cards_points < 17:
                        new_card = ra.choice(self.cards)
                        Dealer_Cards.append(new_card)
                        dealer_cards_points = self.point_calculation(Dealer_Cards)
                        t.sleep(0.5)
                        print(f"Dealer drews: {new_card}")

                    t.sleep(0.5)
                    print(f"Dealer's final cards: {Dealer_Cards} (Total: {dealer_cards_points})")

                    if dealer_cards_points > 21 or player_points > dealer_cards_points:
                        t.sleep(0.5)
                        print("Congratulations! You win!")
                        cash += Dealer_spend
                        t.sleep(0.5)
                        print(f"{cash - Dealer_spend} + {Dealer_spend}$ = You Current Balance is {cash}$")
                        round += 1

                    elif dealer_cards_points > player_points:
                        t.sleep(0.5)
                        print("Dealer wins. Better luck next time!")
                        cash -= spend_cash
                        t.sleep(0.5)
                        print(f"{cash + spend_cash }-{spend_cash} = You Current Balance is {cash}$")
                        round += 1

                    else:
                        t.sleep(0.5)
                        print("It's a Draaaaaaw")
                        t.sleep(0.5)
                        print(f"You Current Balance is {cash - spend_cash}$ + {spend_cash}$ = TOTAL = {cash}")
                        round += 1
                else:
                    t.sleep(0.5)
                    print("You don't have Enough Money Good BYEEEEE")    
                         
                break
            else:
                t.sleep(0.5)
                print("Invalid Number")     


```
